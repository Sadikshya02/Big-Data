```python
data= data.dropna()
```

Data **Visualisation**


Sales by category

```python

import matplotlib.pyplot as plt

data.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.show()
```

Sales by Region


```python
data.groupby("Region")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Region")
plt.ylabel("Sales")
plt.show()
```

**Profit Region**

```python
data["Profit"].hist()
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.show()
```

**Top 10 products by Sales**

```python
top_products = data.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)

top_products.plot(kind="bar")
plt.title("Top 10 Sub-Categories by Sales")
plt.show()
```

**Box Plot**

Box Plot for sales


```python
import matplotlib.pyplot as plt

plt.boxplot(data["Sales"])
plt.title("Box Plot of Sales")
plt.ylabel("Sales")
plt.show()
```

**Box Plot for Profit**

```python
plt.boxplot(data["Profit"])
plt.title("Box Plot of Profit")
plt.ylabel("Profit")
plt.show()
```

**Box Plot for Quantity**


```python
plt.boxplot(data["Quantity"])
plt.title("Box Plot of Quantity")
plt.ylabel("Quantity")
plt.show()
```

**Box Plot for Discount**

```python
plt.boxplot(data["Discount"])
plt.title("Box Plot of Discount")
plt.ylabel("Discount")
plt.show()
```

**Multiple Box Plot Together**

```python
data[["Sales","Profit","Quantity","Discount"]].boxplot()
plt.title("Box Plot of Numerical Features")
plt.show()
```

**Box Plot: Sales by Category**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.boxplot(x="Category", y="Sales", data=data)

plt.title("Box Plot of Sales by Category")
plt.show()
```

**Box Plot: Profit by Category**

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Category", y="Profit", data=data)

plt.title("Box Plot of Profit by Category")
plt.show()
```

**Pie Chart: Sales by Segment**

```python
segment_sales = data.groupby("Segment")["Sales"].sum()

plt.pie(segment_sales, labels=segment_sales.index, autopct="%1.1f%%")
plt.title("Sales Distribution by Segment")
plt.show()
```

Scatter Plot: Sales vs Profit

```python
plt.scatter(data["Sales"], data["Profit"])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit")
plt.show()
```

Line Chart: Sales Trend Over Time

```python
data["Order Date"] = pd.to_datetime(data["Order Date"])
```

```python
sales_trend = data.groupby("Order Date")["Sales"].sum()

sales_trend.plot(kind="line")
plt.title("Sales Trend Over Time")
plt.ylabel("Sales")
plt.show()
```

**Heatmap: Corelation between variables**

```python
import seaborn as sns

corr = data[["Sales","Profit","Quantity","Discount"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
